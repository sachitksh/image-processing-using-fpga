### kernels

```python
sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

blur = np.array([
    [0.0625, 0.125, 0.0625],
    [0.125,  0.25,  0.125],
    [0.0625, 0.125, 0.0625]
])
Ridge_detection = np.array([
    [-1, -1, -1],
    [-1, 4, -1],
    [-1, -1, -1]
])
```

### Convolution Algorithm

```python
def convolve(image,kernel):
    z = np.zeros((np.shape(image)[0]-2,np.shape(image)[1]-2))
    for i in range(np.shape(image)[0]-2):
        for j in range(np.shape(image)[1]-2):
            if(i<np.shape(image)[0]-2 and j<np.shape(image)[1]-2):
                z[i][j]=int((np.multiply(image[i:i+3,[j,j+1,j+2]],kernel)).sum())
    return z
```

```python
#saving image
a=convolve(img,sharpen)
cv2.imwrite('dog2sharpened.jpg',a)
```

```python
img_sharpened = cv2.imread("dog2sharpened.jpg")
```

## sharpened image

```python
_, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img, cmap='gray')
ax[1].imshow(img_sharpened, cmap='gray');
```

```python
#saving image
b=convolve(img,blur)
cv2.imwrite('dog2blured.jpg',b)
```

```python
img_blured = cv2.imread("dog2blured.jpg")
```

## Blured image

```python
_, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img, cmap='gray')
ax[1].imshow(img_blured, cmap='gray');
```
